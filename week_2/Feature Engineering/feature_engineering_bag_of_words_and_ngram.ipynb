## Feature Engineering for Machine Learning - Bag of Words (BOW)

Bag of Words (BoW) is a simple way to represent text data by counting how many times each word appears in a text document, 

while completely ignoring the order and structure of the words.

Think of it as dumping all the words from a document into a "bag," shaking it up, and then counting how many of each word you have.

Simple Analogy:
    
    Imagine you have two sentences:

    - "The cat sat on the mat."

    - "The dog sat on the log."

To compare them, you just list all the unique words and count their frequency, forgetting grammar and word order.

Creating the Bag of Words:

    Vocabulary (all unique words): [the, cat, sat, on, mat, dog, log]

    Vector Representation:

        Sentence 1: [2, 1, 1, 1, 1, 0, 0]

        - (2 "the", 1 "cat", 1 "sat", 1 "on", 1 "mat", 0 "dog", 0 "log")

        Sentence 2: [2, 0, 1, 1, 0, 1, 1]

        - (2 "the", 0 "cat", 1 "sat", 1 "on", 0 "mat", 1 "dog", 1 "log")

Key Point:

BoW converts text into fixed-length numerical vectors based purely on word counts (or presence). 

It's called a "bag" because the order of words is lost.

Why it's used:

Its purpose is to transform unstructured text into a structured, numerical format that machine learning algorithms can understand. 

It's fundamental for:

- Document Classification: Categorizing emails as spam/ham, or news articles by topic.

- Simple Similarity Analysis: Finding how similar two documents are based on their word content.

- As a baseline model: It's a simple, fast starting point before moving to more complex techniques.

Limitation: Since it ignores word order, the sentences "The dog bit the man" and 

"The man bit the dog" would have an identical BoW representation, despite having opposite meanings.

### Data cleaning and preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('stopwords')
```

```python
train_data.columns
```

```python
train_data.drop(columns=['id'], inplace=True)
```

```python
train_data.head()
```

```python
train_data.columns
```

```python
# Define the fraction of the dataset to use (e.g., 10%)
fraction = 0.10 

# Calculate the number of rows to use
num_rows = int(len(train_data) * fraction)

# Select a random sample of rows
train_data = train_data.sample(n=num_rows, random_state=42) 
```

```python
porter_stemmer = PorterStemmer()
processed_word = []

for sentence in train_data.index:
    review = re.sub('[^a-zA-Z]', " ", train_data.loc[sentence, 'review'])
    review = review.lower()
    review = review.split()
    stemmed_word = [porter_stemmer.stem(word) for word in review if word not in set(stopwords.words('english'))]
    processed_word.append(" ".join(stemmed_word))
    
```

```python
processed_word
```

### Create the bag of word model

```python
len(processed_word)
```

```python
#create countVectorizer object
vectorizer = CountVectorizer(max_features=5000)
```

```python
#now get my features for training the model
X = vectorizer.fit_transform(processed_word).toarray()
```

```python
#view the most frequent words in the vocabulary using your vectorizer
vectorizer.vocabulary_
```

## Feature Engineering for Machine Learning - N_Gram


N-grams are contiguous sequences of 'n' items (typically words) from a given text. 

They capture language patterns and context by looking at word combinations rather than single words.

Think of it as a sliding window that moves across the text, capturing phrases of a specific length.

Simple Analogy:

Imagine you're reading text through a magnifying glass that only shows you 2 or 3 words at a time. 

Each little chunk you see through the glass is an n-gram.

Key Point:

The 'n' represents the number of items in the sequence:

- Unigram (1-gram): Single words - "the", "cat", "sat"

- Bigram (2-gram): Two-word phrases - "the cat", "cat sat", "sat on"

- Trigram (3-gram): Three-word phrases - "the cat sat", "cat sat on"

Example:

For the sentence: "The quick brown fox jumps"

- Unigrams: "The", "quick", "brown", "fox", "jumps"

- Bigrams: "The quick", "quick brown", "brown fox", "fox jumps"

- Trigrams: "The quick brown", "quick brown fox", "brown fox jumps"

Why it's used:

N-grams preserve some local word order and context that's lost in Bag of Words. They're essential for:

- Language Modeling: Predicting the next word in a sequence (like smartphone keyboards)

- Improving Text Classification: Capturing meaningful phrases like "not good" instead of just "not" and "good"

- Machine Translation: Understanding common phrases and their translations

- Spell Checking: Identifying common word sequences to detect errors



> Note: N-grams can also be used with characters, not just words, for applications like DNA sequencing or plagiarism detection.

## N_Gram code Implementation 

```python
#create countVectorizer object
vectorizer = CountVectorizer(max_features=100, ngram_range=(2, 3))
```
