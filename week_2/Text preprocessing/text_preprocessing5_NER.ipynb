## Text Preprocessing for Machine Learning - Named Entity Recognition (NER)


Named Entity Recognition (NER) is the automated process of identifying and 

classifying key information (entities) in text into predefined categories.

Think of it as a highlighter that not only finds important words but also labels what type of thing they are.

Simple Analogy:
    Imagine you're reading a news article. NER is the process of circling all the important names and then writing a label above them: PERSON, PLACE, COMPANY, DATE.

Key Point:
    NER doesn't just find words; it understands their semantic type in the context of the sentence.

Examples:

- In the sentence: "Tim Cook announced that Apple will launch a new product in Tokyo next January."

A NER system would identify:

- Tim Cook → PERSON

- Apple → ORGANIZATION

- Tokyo → LOCATION (or GPE - Geo-Political Entity)

- January → DATE

Why it's used:

Its purpose is to extract structured information from unstructured text. This is crucial for:

Search Engines: Improving search by understanding if you're searching for a person, a place, etc.

Content Recommendation: "If you read about Tim Cook, you might like articles about Apple."

Data Analysis: Quickly pulling all company names or locations from thousands of documents.

Customer Support: Automatically routing a ticket by extracting product names and problem types from a customer's message.

```python
import nltk
nltk.download(['maxent_ne_chunker', 'punkt_tab', 'averaged_perceptron_tagger_eng', 'maxent_ne_chunker_tab', 'words'])
from nltk import word_tokenize, pos_tag, ne_chunk , ne_chunk
```

```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, whose company specialized in building metal frameworks and structures"

tokenized_word = word_tokenize(sentence)
tokenized_word 
```

```python
tagged_word = pos_tag(tokenized_word)

tagged_word
```
