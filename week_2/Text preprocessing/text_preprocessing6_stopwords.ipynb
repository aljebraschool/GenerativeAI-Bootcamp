## Text Preprocessing for Machine Learning - Stopwords


Stop Words are common words that are often filtered out from text before processing because 

they appear very frequently and carry little meaningful information on their own.

Think of it as removing the grammatical "filler" words to focus on the key content.

Simple Analogy:
    Imagine you're looking for the main topics in a conversation. You would ignore words like "the," "and," "is," and "to" to focus on the important nouns and verbs.

Key Point:
    The goal is to reduce noise and improve efficiency by focusing computational resources on the words that truly define the meaning of a text.

Examples of Common Stop Words:

- Articles: a, an, the

- Conjunctions: and, but, or

- Prepositions: in, on, at, to, for

- Pronouns: I, you, he, she, it

- Common Verbs: is, are, was, be, have

In a sentence:
"The cat is sitting on the mat and looking at the window."

After removing stop words, you're left with the key terms:
"cat", "sitting", "mat", "looking", "window"

Why it's used:

Its purpose is to improve performance and focus in:

Search Engines: Indexing and matching queries based on meaningful keywords.

Text Analysis & Modeling: Improving the quality of data for tasks like sentiment analysis or topic modeling by removing linguistic clutter.

Efficiency: Reducing the overall size and complexity of the data being processed.

> Note: The specific list of stop words can vary depending on the application and language. Sometimes, for tasks like phrase search or language translation, these words are crucial and are not removed.

## stopwords code Implementation 

```python
paragraph = """
I have three visions for India. In 3000 years of our history, people from all over  
the world have come and invaded us, captured our lands, conquered our minds.  
From Alexander onwards, the Greeks, the Turks, the Mughals, the Portuguese, the British,  
the French, the Dutch, all of them came and looted us, took over what was ours.  
Yet we have not done this to any other nation. We have not conquered anyone.  
We have not grabbed their land, their culture, their history and tried to enforce our way of life on them.  
Why? Because we respect the freedom of others. That is why my  
first vision is that of freedom. I believe that India got its first vision of  
this in 1857, when we started the War of Independence. It is this freedom that  
we must protect and nurture and build on. If we are not free, no one will respect us.  

My second vision for India’s development. For fifty years we have been a developing nation.  
It is time we see ourselves as a developed nation. We are among the top 5 nations of the world  
in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.  
Our achievements are being globally recognized today. Yet we lack the self-confidence to  
see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?  

I have a third vision. India must stand up to the world. Because I believe that unless India  
stands up to the world, no one will respect us. Only strength respects strength. We must be  
strong not only as a military power but also as an economic power. Both must go hand-in-hand.  

My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of  
Space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.  
I was lucky to have worked with all three of them closely and consider this the great opportunity of my life.  
I see four milestones in my career.  

"""
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
from nltk import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
```

### See some available stopwords in english (first ten words)

```python

stopwords.words('english')[:10]
```

### Tokenize the paragraph to sentences

```python
tokenized_sentences = nltk.sent_tokenize(paragraph)

tokenized_sentences[:10]
```

### Now check words in each sentence (above) that are not in english stopwords, then apply Porter stemming

```python
porter_stemmer = PorterStemmer()
processed_word = []

for sentence in tokenized_sentences:
     #tokenized each sentence into word 
    tokenized_word = word_tokenize(sentence)
    #apply stemming to only the words not found in english stopwords
    stemmed_word = [porter_stemmer.stem(word) for word in tokenized_word if word not in set(stopwords.words('english'))]
    #put the joint tokenized words in each location of the tokenized words above to get sentences
    processed_word.append(" ".join(stemmed_word))
```

### Lemmatization 

```python
from nltk.stem import WordNetLemmatizer

lemmatization = WordNetLemmatizer()
```

```python
processed_word = []

for sentence in tokenized_sentences:
     #tokenized each sentence into word 
    tokenized_word = word_tokenize(sentence)
    #apply stemming to only the words not found in english stopwords
    stemmed_word = [lemmatization.lemmatize(word.lower()) for word in tokenized_word if word not in set(stopwords.words('english'))]
    #put the joint tokenized words in each location of the tokenized words above to get sentences
    processed_word.append(" ".join(stemmed_word))
```
