```python
import nltk
nltk.download(['maxent_ne_chunker', 'punkt_tab', 'averaged_perceptron_tagger_eng', 'maxent_ne_chunker_tab', 'words'])
from nltk import word_tokenize, ne_chunk, pos_tag 
```

```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, \
whose company specialized in building metal frameworks and structures"

tokenize = word_tokenize(sentence)

tokenize

```

```python
tagged_word = pos_tag(tokenize)

tagged_word
```

```python
ne_chunk(tagged_word)
```
